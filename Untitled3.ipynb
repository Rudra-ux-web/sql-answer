# Question 1: Explain the fundamental differences between DDL, DML, and DQL commands in SQL. Provide one example for each type of command.

## Answer

SQL (Structured Query Language) is used to create, manage, and retrieve data from databases. SQL commands are divided into different categories based on the work they perform. Among them, DDL, DML, and DQL are the most commonly used commands.

### DDL (Data Definition Language)

DDL commands are used to define the structure of a database. These commands help us create, modify, or delete database objects such as tables. In simple words, DDL is used when we want to design the database structure.

Some common DDL commands are CREATE, ALTER, DROP, and TRUNCATE.

Example:

```sql
CREATE TABLE Students (
    StudentID INT PRIMARY KEY,
    StudentName VARCHAR(100)
);
```

The above command creates a new table named Students with two columns.

### DML (Data Manipulation Language)

DML commands are used to work with the data stored inside tables. These commands allow us to add, update, or remove records from a table.

Some common DML commands are INSERT, UPDATE, and DELETE.

Example:

```sql
INSERT INTO Students (StudentID, StudentName)
VALUES (1, 'John');
```

This command inserts a new record into the Students table.

### DQL (Data Query Language)

DQL commands are used to retrieve data from the database. The SELECT statement is the most commonly used DQL command. It helps users view and analyze stored information.

Example:

```sql
SELECT * FROM Students;
```

This query displays all records from the Students table.

### Difference Between DDL, DML, and DQL

DDL is used to create and manage database structures, DML is used to modify the data inside those structures, and DQL is used to retrieve information from the database. For example, if we create a table using CREATE, it is a DDL operation. If we insert records into the table, it becomes a DML operation. When we view those records using SELECT, it is a DQL operation.

### Conclusion

DDL, DML, and DQL are important parts of SQL. DDL helps in creating database structures, DML helps in managing data, and DQL helps in retrieving data. Together they make database management easier and more efficient.


# Question 2: What is the purpose of SQL constraints? Name and describe three common types of constraints, providing a simple scenario where each would be useful.

## Answer

SQL constraints are rules that are applied to table columns to ensure that only valid and accurate data is stored in the database. They help maintain data integrity and prevent users from entering incorrect or duplicate information. Constraints are very important because they improve the reliability and consistency of the database.

Without constraints, a database may contain duplicate records, missing values, or invalid data, which can create problems during data analysis and reporting.

### 1. PRIMARY KEY Constraint

A PRIMARY KEY is used to uniquely identify each record in a table. It does not allow duplicate values or NULL values.

Example:

```sql
CustomerID INT PRIMARY KEY
```

Scenario:

Suppose a company maintains customer information. Each customer must have a unique CustomerID so that records can be identified correctly. If two customers have the same ID, it would create confusion.

### 2. NOT NULL Constraint

The NOT NULL constraint ensures that a column cannot have an empty value.

Example:

```sql
ProductName VARCHAR(100) NOT NULL
```

Scenario:

In an online shopping website, every product must have a name. If the product name is left blank, customers will not know what product they are buying.

### 3. UNIQUE Constraint

The UNIQUE constraint ensures that all values in a column are different from each other.

Example:

```sql
Email VARCHAR(100) UNIQUE
```

Scenario:

When users register on a website, each email address should be unique. Two users should not be allowed to register using the same email ID.

### Conclusion

SQL constraints help maintain data quality and enforce important business rules. They prevent invalid data from entering the database and make database management more efficient and reliable.


# Question 3: Explain the difference between LIMIT and OFFSET clauses in SQL. How would you use them together to retrieve the third page of results, assuming each page has 10 records?

## Answer

When working with large amounts of data, it is often not practical to display all records at once. SQL provides LIMIT and OFFSET clauses to control how many records are displayed and from where the records should start.

### LIMIT Clause

The LIMIT clause is used to specify the maximum number of rows that should be returned by a query.

Example:

```sql
SELECT * FROM Products
LIMIT 10;
```

The above query returns only the first 10 records from the Products table.

### OFFSET Clause

The OFFSET clause is used to skip a specific number of rows before displaying the result.

Example:

```sql
SELECT * FROM Products
LIMIT 10 OFFSET 10;
```

This query skips the first 10 rows and displays the next 10 rows.

### Difference Between LIMIT and OFFSET

LIMIT controls how many rows will be displayed, whereas OFFSET controls how many rows will be skipped before displaying the results.

### Using LIMIT and OFFSET for Pagination

Pagination is commonly used in websites where data is displayed page by page.

Assume each page contains 10 records.

* Page 1 → OFFSET 0
* Page 2 → OFFSET 10
* Page 3 → OFFSET 20

To display the third page, we need to skip the first 20 records and display the next 10 records.

SQL Query:

```sql
SELECT *
FROM Products
LIMIT 10 OFFSET 20;
```

### Conclusion

LIMIT and OFFSET are very useful for pagination and efficient data retrieval. They help improve application performance and provide a better user experience when working with large datasets.


# Question 4: What is a Common Table Expression (CTE) in SQL, and what are its main benefits? Provide a simple SQL example demonstrating its usage.

## Answer

A Common Table Expression (CTE) is a temporary result set that is created within a SQL query using the WITH keyword. It exists only during the execution of that query and is not permanently stored in the database.

CTEs are mainly used to simplify complex queries and make SQL code easier to read and understand.

### Syntax of CTE

```sql
WITH CTE_Name AS
(
    SELECT column_name
    FROM table_name
)
SELECT *
FROM CTE_Name;
```

### Example

Suppose we want to find products whose price is greater than 100.

```sql
WITH ExpensiveProducts AS
(
    SELECT ProductName, Price
    FROM Products
    WHERE Price > 100
)
SELECT *
FROM ExpensiveProducts;
```

### Explanation

In this example, a CTE named ExpensiveProducts is created. It stores products having a price greater than 100. The main query then retrieves data from the CTE.

### Benefits of CTE

1. Improves readability of complex SQL queries.
2. Makes queries easier to maintain.
3. Reduces repetition of code.
4. Helps break large queries into smaller logical parts.
5. Supports recursive queries for hierarchical data.

### CTE vs Subquery

Although both CTEs and subqueries can produce temporary results, CTEs are generally easier to read and understand, especially when dealing with large and complex SQL statements.

### Conclusion

CTEs are a powerful SQL feature that improves query readability and simplifies complex data analysis tasks. They are widely used in reporting and business intelligence applications.


# Question 5: Describe the concept of SQL Normalization and its primary goals. Briefly explain the first three normal forms (1NF, 2NF, 3NF).

## Answer

Normalization is a database design technique used to organize data in a proper way so that data redundancy and inconsistency can be reduced. In simple words, normalization helps us store data efficiently by dividing large tables into smaller related tables.

The main objective of normalization is to avoid duplicate data and maintain data integrity. It also makes the database easier to maintain and update.

### Goals of Normalization

The primary goals of normalization are:

1. To reduce data redundancy.
2. To eliminate duplicate data.
3. To improve data consistency.
4. To maintain data integrity.
5. To make database maintenance easier.

### First Normal Form (1NF)

A table is said to be in First Normal Form if:

* Each column contains only atomic (single) values.
* There are no repeating groups.
* Each row is unique.

Example:

Not in 1NF:

| StudentID | Subjects      |
| --------- | ------------- |
| 1         | Math, Science |

In 1NF:

| StudentID | Subject |
| --------- | ------- |
| 1         | Math    |
| 1         | Science |

Here, each cell contains only one value.

### Second Normal Form (2NF)

A table is in Second Normal Form if:

* It is already in 1NF.
* All non-key attributes depend on the entire primary key.

Example:

Instead of storing student and course information in one table, separate tables can be created for Students, Courses, and Enrollments.

This removes partial dependency and reduces redundancy.

### Third Normal Form (3NF)

A table is in Third Normal Form if:

* It is already in 2NF.
* There are no transitive dependencies.

Example:

Not in 3NF:

| EmployeeID | DepartmentID | DepartmentName |

In 3NF:

Employee Table:

| EmployeeID | DepartmentID |

Department Table:

| DepartmentID | DepartmentName |

This removes unnecessary dependencies between non-key attributes.

### Advantages of Normalization

* Reduces duplicate data.
* Improves database performance.
* Makes updates easier.
* Improves consistency and accuracy.
* Saves storage space.

### Conclusion

Normalization is an important concept in database design. The first three normal forms (1NF, 2NF, and 3NF) help organize data efficiently and reduce redundancy, making the database more reliable and easier to manage.

````python
sql_query_cte = """
WITH RankedProducts AS
(
    SELECT
        c.CategoryName,
        p.ProductName,
        p.Price,
        ROW_NUMBER() OVER
        (
            PARTITION BY c.CategoryName
            ORDER BY p.Price DESC
        ) AS RankNo
    FROM Products p
    INNER JOIN Categories c
    ON p.CategoryID = c.CategoryID
)

SELECT
    CategoryName,
    ProductName,
    Price
FROM RankedProducts
WHERE RankNo <= 2
ORDER BY CategoryName, Price DESC;
"""

print("\n--- Top 2 Products per Category ---")
print(pd.read_sql_query(sql_query_cte, conn))
````

## 1. Identify the Top 5 Customers Based on Total Amount Spent

```sql
SELECT
    CONCAT(c.first_name,' ',c.last_name) AS CustomerName,
    c.email,
    SUM(p.amount) AS TotalAmountSpent
FROM customer c
JOIN payment p
ON c.customer_id = p.customer_id
GROUP BY c.customer_id, c.first_name, c.last_name, c.email
ORDER BY TotalAmountSpent DESC
LIMIT 5;
```

### Explanation

In this query, I joined the Customer and Payment tables using the customer_id column. The SUM() function is used to calculate the total amount spent by each customer. The GROUP BY clause groups the records customer-wise, while ORDER BY arranges the results in descending order based on total spending. Finally, LIMIT 5 displays only the top five customers who have spent the highest amount.

## 2. Find the Top 3 Movie Categories with the Highest Rental Counts

```sql
SELECT
    cat.name AS CategoryName,
    COUNT(*) AS RentalCount
FROM rental r
JOIN inventory i
ON r.inventory_id = i.inventory_id
JOIN film_category fc
ON i.film_id = fc.film_id
JOIN category cat
ON fc.category_id = cat.category_id
GROUP BY cat.name
ORDER BY RentalCount DESC
LIMIT 3;
```

### Explanation

This query identifies the most popular movie categories based on rental frequency. Multiple tables are joined to connect rental records with movie categories. The COUNT() function calculates the total number of rentals for each category. The results are then sorted in descending order and only the top three categories are displayed.

## 3. Calculate How Many Films Are Available at Each Store and How Many Have Never Been Rented

```sql

            ELSE 0
        END

JOIN inventory i
ON s.store_id = i.store_id

ON i.inventory_id = r.inventory_id
GROUP BY s.store_id;
```

### Explanation

## 4. Show the Total Revenue Per Month for the Year 2023

```sql
SELECT
    MONTH(payment_date) AS MonthNumber,
    SUM(amount) AS TotalRevenue
FROM payment
WHERE YEAR(payment_date) = 2023
GROUP BY MONTH(payment_date)
ORDER BY MonthNumber;
```

### Explanation

This query calculates the total revenue earned during each month of the year 2023. The MONTH() function extracts the month from the payment date, while the SUM() function calculates the total revenue generated during that month. The GROUP BY clause groups records month-wise, and ORDER BY ensures that the months are displayed in chronological order. This analysis helps identify seasonal trends in the business.

## 5. Identify Customers Who Have Rented More Than 10 Times in the Last 6 Months

```sql
SELECT
    c.customer_id,
    CONCAT(c.first_name,' ',c.last_name) AS CustomerName,
    COUNT(r.rental_id) AS TotalRentals
FROM customer c
JOIN rental r
ON c.customer_id = r.customer_id
WHERE r.rental_date >= DATE_SUB(CURDATE(), INTERVAL 6 MONTH)
GROUP BY c.customer_id,
         c.first_name,
         c.last_name
HAVING COUNT(r.rental_id) > 10
ORDER BY TotalRentals DESC;
```

### Explanation

This query helps identify loyal customers who frequently rent movies. The Customer and Rental tables are joined using customer_id. The WHERE clause filters rental records from the last six months. The COUNT() function calculates the total number of rentals for each customer. The HAVING clause is then used to display only customers who have rented more than ten times. Such customers can be targeted for loyalty programs and special offers.
